##### 定义使用模型训练的函数

```python
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix, f1_score
```

```python
def model_train(model, x_train, y_train, x_test, y_test, name = 'Fraud'):
    model = model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_pred,y_test)
    recall = recall_score(y_pred,y_test)
    auc = roc_auc_score(y_pred, y_test)
    f1 = f1_score(y_pred,y_test)
    confusion = confusion_matrix(y_pred, y_test)
    print('Model Used: ', model)
    print('{} accuracy is {}%'.format(name, accuracy*100))
    print('{} recall is {}%'.format(name, recall*100))
    print('{} auc is {}%'.format(name, auc*100))
    print('{} f1 is {}%'.format(name, f1*100))
    print('{} confusion is \n{}'.format(name, confusion))
    return model
```

```python
#def model_train(model, x_train, y_train, x_test, y_test, name = 'Fraud')
model = model_train(model_fraud,x_fraud_train,y_fraud_train,x_fraud_test,y_fraud_test,'Fraud')
```

```python
#def model_train(model, x_train, y_train, x_test, y_test, name = 'Fraud')
model = model_train(model_late,x_late_train,y_late_train,x_late_test,y_late_test,'Late')
```

##### 线性回归模型

```python
from sklearn.metrics import explained_variance_score, mean_squared_error
```

```python
def model_train(model, x_train, y_train, x_test, y_test, name):
    model = model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    var = explained_variance_score(y_pred, y_test)
    mse = mean_squared_error(y_pred,y_test)
    print('Model Used: ', model)
    print('{} variance is {}'.format(name, var))
    print('{} mse is {}'.format(name, mse))
    print('ypred: \n', y_pred)
    print('ytest: \n', y_test)
    return model
```

```python
from sklearn.linear_model import LinearRegression
```

##### 对销售额度进行预测并返回模型的explained variance和mse

```python
model_sales = LinearRegression()
```

```python
model = model_train(model_sales,x_sales_train,y_sales_train,x_sales_test,y_sales_test,'Sales')
```

##### 对订货数量进行预测并返回模型的explained variance和mse

```python
model_quantity = LinearRegression()
```

##### 通过打印 ypred和 ytest可以看出，使用线性回归预测订货数量之所以mse比较小是因为订货数量都是整数而且取值范围很小，基本只有1和2这两个数字，所以预测效果好

```python
model = model_train(model_quantity,x_quantity_train,y_quantity_train,x_quantity_test,y_quantity_test,'Quantity')
```

##### 线性回归模型的可解释性比较高，比如物品的分类对销售额影响比较高，比如钓鱼用具因为普遍造价高所以比通用的运动物品销售额高
